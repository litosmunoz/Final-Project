```python
import pandas as pd
```

```python
df = pd.read_csv ('./dataset/unemployment_rate.csv', encoding='unicode_escape')
```

```python
df['time'] = df['time'].str.extract('(\d{4}-\d{2}-\d{2})', expand=True)
df['time'] = pd.to_datetime(df["time"])
df.set_index('time', inplace=True)
```

```python
df_reindexed = df.reindex(pd.date_range(start=df.index.min(),
                                                  end=df.index.max(),
                                                  freq='1D'))
```

```python
df_reindexed
```

```python
df_reindexed.interpolate(method='pad', inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config Inlinebackend.figure_format= 'retina'
sns.set_context("poster")
sns.set(rc={"figure.figsize": (12.,6.)})
sns.set_style("white")
```

```python
plt.plot(df_reindexed, color= 'b')
plt.title('Unemployment Rate', fontsize= 20)
plt.xlabel('Date', fontsize= 20)
plt.ylabel('Unemployment Rate', fontsize= 20)
plt.style.use('seaborn')
```
